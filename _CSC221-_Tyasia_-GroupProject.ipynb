```python
import ssl 
ssl._create_default_https_context = ssl._create_unverified_context

import urllib.request 
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_population'

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
req = urllib.request.Request(url, headers=headers)

page = urllib.request.urlopen(req)

soup = BeautifulSoup(page, "lxml")

print(soup.prettify())

print("\nTitle (with tags):")
print(soup.title)

print("\nTitle (without tags):")
print(soup.title.string)

```

```python
all_tables = soup.find_all("table")

all_sortable_tables = soup.find_all("table", class_="wikitable sortable")

# The population ranking table is sortable_tables[0]
right_table = all_sortable_tables[0]

print("RIGHT TABLE:\n")
print(right_table.prettify())
```

```python
A = []
B = []
C = []
D = []
E = []

for row in right_table.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) >= 5:
        A.append(cells[0].get_text(strip=True))
        B.append(cells[1].get_text(strip=True))
        C.append(cells[2].get_text(strip=True))
        D.append(cells[3].get_text(strip=True))
        E.append(cells[4].get_text(strip=True))
        
print(A[:10])
print(B[:10])
print(C[:10])
print(D[:10])
print(E[:10])
```

```python
df = pd.DataFrame(A, columns=['Rank'])
df['State'] = B
df['Census_2020'] = C
df['Estimate_2023'] = D
df['Percent_Change'] = E

pd.set_option('display.max_rows', None)

df
```
